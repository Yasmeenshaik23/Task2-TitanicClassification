```python
df=df.replace({'Sex':{'male':1,'female':0}, 'Embarked':{'S':0,'C':1,'Q':3}})
```

```python
x=df.drop(columns=['PassengerId','Name','Ticket','Survived'],axis=1)
y=df['Survived']
```

```python
x
```

```python
y
```

```python
x_tr,x_test,y_tr,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_tr.shape,x_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_tr, y_tr)
```

```python
x_tr_prediction=model.predict(x_tr)
```

```python
print(x_tr_prediction)
```

```python
training_data_accuracy = accuracy_score(y_tr,x_tr_prediction)
print('Accuracy score of training data : ',training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
```

```python
print(x_test_prediction)
```

```python
test_data_accuracy = accuracy_score(y_test,x_test_prediction)
print('Accuracy score of test data : ',test_data_accuracy)
```
